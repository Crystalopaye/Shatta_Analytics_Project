### Cleaning the songs_fact dataframe

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load Data
df = pd.read_csv("C:/fiifi/Shatta_Analytics_Project/CSV_files/csv_files_utf8/songs_fact.csv")

# Data Cleaning
df = df.drop(columns="notes", axis=1)  # Drop notes column in df
df = df.iloc[:969]   # drop all NaN rows in df
```

```python
# Extract date part
df['date_part'] = df['release_date'].astype(str).str.split(' ').str[0]

# Convert: if it has '-', use standard format; if '/', use dayfirst
df['release_datetime'] = df['date_part'].apply(
    lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='coerce') if pd.notna(x) and '-' in str(x) 
    else pd.to_datetime(x, dayfirst=True, errors='coerce') if pd.notna(x) and '/' in str(x)
    else pd.NaT
)

# Drop date_part
df = df.drop(columns="date_part")

# Rename release_date to release_date_original & release_datetime to release_date
df = df.rename(columns={"release_date": "release_date_original",
                   "release_datetime": "release_date"})
```

```python
# export df to csv.
df.to_csv("songs_fact.csv")
```
